## Section 2 - getDiceGuess()

### Original (buggy) code:
```python
def getDiceGuess():
    guesss = 0
    while guesss not in ('1 ', '2 ', '3 ', '4 '):
        print('Guess the number on a 4-sided dice! Enter a number between 1 and 4:')
        guess = input()
    return guesss
```

### Bugs fixed:
| Fix | Description |
|-----|-------------|
| Fix 1 | Trailing spaces in `('1 ', '2 ', '3 ', '4 ')` - removed so input can actually match |
| Fix 2 & 3 | Variable named `guesss` (3 s) - renamed to `guess` throughout |
| Fix 4 | `return guesss` returned 0 - changed to `return int(guess)` so type matches dice roll |

````python
def getDiceGuess():
    guess = 0
    while guess not in ('1', '2', '3', '4'): # Fix 1: removed trailing spaces from the tuple
        print('Guess the number on a 4-sided dice! Enter a number between 1 and 4:')
        guess = input() # Fix 2 & 3: was 'guesss' (3 s), renamed to 'guess'
    return int(guess) # Fix 4: was 'return guesss' returning 0, now returns int(guess)
````

## Section 3 - rollDice()

### Original (buggy) code:
```python
def rollDice():
    dice = random.randint(1, 5)
    if dice == guess:
        print('You are good!')
    else:
```

### Bugs fixed:
| Fix | Description |
|-----|-------------|
| Fix 5 | `random.randint(1, 5)` - changed `random` to `rnd` (the alias) and range to 1–4 (4-sided dice has no 5) |
| Fix 6 | `if dice == guess` used undefined global `guess` - replaced with `getDiceGuess()` call |
| Fix 7 | `if dice = guess` (assignment) - changed to `==` (comparison) |
| Fix 8 | `guesss = input()` (wrong name, raw input) - replaced with `getDiceGuess()` |
| Fix 9 | `if dice = guess` - changed to `==` |
| Fix 10 | `guess = input()` (raw unvalidated input) - replaced with `getDiceGuess()` |
| Fix 11 | `if dice = guess` - changed to `==` |
| Fix 12 | `guesss = input()` (wrong name, raw input) - replaced with `getDiceGuess()` |
| Fix 13 | `if dice = guess` - missing colon and using `=`, changed to `==` and added `:` |

````python
def rollDice():
    dice = rnd.randint(1, 4) # Fix 5: was 'random.randint(1, 5)' - changed alias and range

    guess = getDiceGuess() # Fix 6: was using undefined global 'guess' - now calls getDiceGuess()
    if dice == guess: # Fix 7: was 'if dice = guess' - changed to ==
        print('You got it on the first try!')
    else:
        print('Sorry! Try to guess again!')

        guess = getDiceGuess() # Fix 8: was 'guesss = input()' - replaced with getDiceGuess()
        if dice == guess: # Fix 9: was 'if dice = guess' - changed to ==
            print('Congrats! You got it!')
        else:
            print('Sorry! Your third chance!')

            guess = getDiceGuess() # Fix 10: was 'guess = input()' - replaced with getDiceGuess()
            if dice == guess: # Fix 11: was 'if dice = guess' - changed to ==
                print('Congrats! You got it!')
            else:
                print('Sorry! You still have one try left!')

                guess = getDiceGuess() # Fix 12: was 'guesss = input()' - replaced with getDiceGuess()
                if dice == guess: # Fix 13: was 'if dice = guess' missing colon - fixed to == and added ':'
                    print('Congrats! You got it!')
                else:
                    print('Nope. You are really bad at this game.')
                    print(f'The number was {dice}.')
````
