# Task 1 - Supernacci Numbers
### Computing the n-th number in the Supernacci sequence
**Method:** List-based iteration using slicing and sum() - no recursion

The Supernacci sequence is defined as:
- `s(1) = 1`
- `s(n)` is the sum of the last `(n - 1) % 3` Supernacci numbers
- If the sum is empty, then `s(n) = s(n - 1)`

```
def supernacci(n):
    """
    Computes and returns the n-th Supernacci number.

    Parameters:
        n (int): The position in the Supernacci sequence.

    Returns:
        int: The n-th Supernacci number.
    """

    # Check that n is valid
    if n < 1:
        raise ValueError("n must be at least 1")

    # The list is made 1-indexed by placing a dummy value at index 0.
    # This means that s[1] represents s(1).
    s = [0, 1]

    # Compute Supernacci numbers from s(2) up to s(n)
    for i in range(2, n + 1):

        # k decides how many of the previous numbers should be summed
        k = (i - 1) % 3

        # If k is 0, the sum is empty.
        # According to the rule, we then use the previous Supernacci number.
        if k == 0:
            s.append(s[-1])

        # Otherwise, sum the last k numbers using slicing and sum()
        else:
            s.append(sum(s[-k:]))

    # Return the n-th Supernacci number
    return s[n]

for i in range(1, 16):
    print(f"s({i}) = {supernacci(i)}")
```
